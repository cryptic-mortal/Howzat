# 2008-2011

```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("ipl_da_08_11.csv")
```

```python
data
```

```python
data[data.isna().any(axis=1)==1]
```

```python
data=data.dropna()
```

```python
data.columns
```

```python
data.drop(["Unnamed: 0"],axis=1,inplace=True)
```

```python
data.info()
```

```python
data.Match.value_counts()
```

```python
season = [2011]
m_type = ["Qualifier 1", "Eliminator", "Qualifier 2", "Final"]
```

```python
data[data.Match=="Final"]
season_1=[2008,2009,2010]
ses_8_9=["SemiFinal_1","SemiFinal_2","Final"]
M_type=ses_8_9 + m_type
```

```python
data
```

```python
data[data.Match=="Final"]
```

```python
for ses in season_1:
  i=0
  m_idx=data.loc[(data["Season"]==ses)&(data["Match"].str.contains("Final"))].index
  for idx in m_idx:
    if i < len(ses_8_9):
            data.at[idx, "Match"] = ses_8_9[i]
            i += 1
```

```python
for ses in season:
  i=0
  m_idx=data.loc[(data["Season"]==ses)&(data["Match"].str.contains("Final"))].index
  for idx in m_idx:
    if i < len(m_type):
            data.at[idx, "Match"] = m_type[i]
            i += 1
```

```python
data.drop("index",axis=1,inplace=True)
```

```python
data
```

```python
for j in M_type:
  data.loc[data["Match"].str.contains(j,na=False),"Match Type"]=j
```

```python
data
```

```python
data["Target_Runs"]=data["Target_Runs"].astype("int")
data["Win_Margin"]=data["Win_Margin"].astype("int")
data.info()
```

```python
data.to_csv("ipl_da_08_11.csv")
```
