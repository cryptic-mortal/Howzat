# 2008-2011

```python
import pandas as pd
import numpy as np
```

```python
data1=pd.read_csv("ipl_da_08_11.csv")
```

```python
data1
```

```python
data1[data1.isna().any(axis=1)==1]
```

```python
data1.loc[68,'Winner']='RR'
data1.loc[68,'Win_Margin']=0
data1.loc[68,'Result_Type']="runs"
data1.loc[68,'Target_Runs']=  150

data1.loc[133,'Winner']="KXIP"
data1.loc[133,'Win_Margin']=0
data1.loc[133,'Result_Type']="runs"
data1.loc[133,'Target_Runs']=  136
```

```python
data1=data1.dropna()
```

```python
data1.columns
```

```python
data1.drop(["Unnamed: 0"],axis=1,inplace=True)
```

```python
data1.info()
```

```python
data1.Match.value_counts()
```

```python
season = [2011]
m_type = ["Qualifier 1", "Eliminator", "Qualifier 2", "Final"]
```

```python
data1[data1.Match=="Final"]
season_1=[2008,2009,2010]
ses_8_9=["SemiFinal_1","SemiFinal_2","Final"]
M_type=ses_8_9 + m_type
```

```python
data1
```

```python
data1[data1.Match=="Final"]
```

```python
for ses in season_1:
  i=0
  m_idx=data1.loc[(data1["Season"]==ses)&(data1["Match"].str.contains("Final"))].index
  for idx in m_idx:
    if i < len(ses_8_9):
            data1.at[idx, "Match"] = ses_8_9[i]
            i += 1
```

```python
for ses in season:
  i=0
  m_idx=data1.loc[(data1["Season"]==ses)&(data1["Match"].str.contains("Final"))].index
  for idx in m_idx:
    if i < len(m_type):
            data1.at[idx, "Match"] = m_type[i]
            i += 1
```

```python
data1.drop("index",axis=1,inplace=True)
```

```python
data1
```

```python
for j in M_type:
  data1.loc[data1["Match"].str.contains(j,na=False),"Match Type"]=j
```

```python
data1
```

```python
data1["Target_Runs"]=data1["Target_Runs"].astype("int")
data1["Win_Margin"]=data1["Win_Margin"].astype("int")
data1.info()
```

```python
data1.to_csv("ipl_da_08_11.csv")
```
